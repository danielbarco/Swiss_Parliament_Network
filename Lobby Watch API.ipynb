```python
import json
import urllib
```

```python
#calling lobbywatch API according to https://github.com/lobbywatch/lobbywatch/blob/master/drupal/lobbywatch/lobbywatch_data/data_interface_description.md

url = 'https://cms.lobbywatch.ch/de/data/interface/v1/json/table/organisation/flat/list'
response = json.loads(urllib.request.urlopen(url).read())
```

```python
#inspect result
print(response)
```

```python
#show interesting stuff: organisation and lobby = interessengruppe
for element in response['data']:
    print(element['anzeige_name']+" : " + element['interessengruppe'])
```

```python
#results seem limited? --> The default is 10. --> limit=none

url = 'https://cms.lobbywatch.ch/de/data/interface/v1/json/table/organisation/flat/list?limit=none'
response = json.loads(urllib.request.urlopen(url).read())
#show interesting stuff: organisation and lobby = interessengruppe
for element in response['data']:
    print(element['anzeige_name']+" : " + element['interessengruppe'])
```

```python
#type error: need to stringitize the response

for element in response['data']:
    print(str(element['anzeige_name'])+" : " + str(element['interessengruppe']))
```

```python
#get lobby for value by example 

value= "comparis.ch AG"

for element in response['data']:
    if element['anzeige_name'] == value:
        print(str(value) + ": Diese Organisation ist in folgender Lobby:  " + str(element['interessengruppe']))
    else:
        pass
```
